# WORLD HAPPINESS RECORD ANALYSIS 2021

This project focuses on analyzing the World Happiness Report 2021 dataset to understand the factors that influence happiness levels across different countries. The dataset includes various indicators such as GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption.

The objective of this project is to explore how these factors contribute to a country's overall happiness score and to identify patterns or trends among the happiest and least happy nations. Using data analysis techniques, the project highlights key insights, comparisons between countries, and the impact of socio-economic conditions on well-being.

Through visualization and statistical analysis, this project aims to provide a better understanding of global happiness and the elements that play a crucial role in improving quality of life.
